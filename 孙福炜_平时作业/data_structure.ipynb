# linked list

```python
class LNode:
    def __init__(self, value, next_):
        self.value = value
        self.next_ = next_
```

```python
def init_linked_list(input_list):
    if not input_list:
        return None
    else:
        return LNode(input_list[0], init_linked_list(input_list[1:]))
```

```python
l = init_linked_list([1, 3, 2, 4, 5])
```

```python
lolviz.objviz(l)
```

```python
def insert_linked_list_sorted(value, l):
    if not l:
        return LNode(value, None)
    elif value < l.value:
        return LNode(value, l)
    elif value == l.value:
        return l
    else:
        return LNode(l.value, insert_linked_list_sorted(value, l.next_))
```

```python
sorted_list = None
for i in [1, 3, 2, 4, 5]:
    sorted_list = insert_linked_list_sorted(i, sorted_list)
```

```python
lolviz.objviz(sorted_list)
```

```python
def reverse_list(l):
    if not l or not l.next_:
        return l
    else:
        ret = reverse_list(l.next_)
        l.next_.next_ = l
        l.next_ = None
        return ret
```

```python
reversed_list = reverse_list(sorted_list)
```

```python
lolviz.objviz(reversed_list)
```

```python
lolviz.objviz(sorted_list)
```

```python
def reverse_list_not_inplace(l):
    if not l:
        return None, None
    elif not l.next_:
        node = LNode(l.value, None)
        return node, node
    else:
        node = LNode(l.value, None)
        head, tail = reverse_list_not_inplace(l.next_)
        tail.next_ = node
        return head, node
```

```python
l = init_linked_list([1, 3, 2, 4, 5])
r = reverse_list_not_inplace(l)
```

```python
lolviz.objviz(r)
```

```python
lolviz.objviz(l)
```

```python
class TNode:
    def __init__(self, value, left, right):
        self.value = value
        self.left = left
        self.right = right
```

```python
def insert_tree(value, node):
    if not node:
        return TNode(value, None, None)
    elif value == node.value:
        return node
    elif value < node.value:
        return TNode(node.value, insert_tree(value, node.left), node.right)
    else:
        return TNode(node.value, node.left, insert_tree(value, node.right))
```

```python
tree = insert_tree(5, insert_tree(3, insert_tree(4, None)))
```

```python
lolviz.treeviz(tree)
```
